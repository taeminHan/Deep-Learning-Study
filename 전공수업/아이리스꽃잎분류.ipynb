```python
import tensorflow as tf
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
# iris.data = 각각의 꽃의 꽃잎의 넓이, 길이 등의 데이터
# iris.data = 해당 꽃들의 품종 이름 0 = setosa, 1 = versicolor, 2 = virginica
X = iris.data
y = iris.target
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
enc = OneHotEncoder()
y_onehot = enc.fit_transform(y.reshape(-1, 1)).toarray()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(64, activation='relu'))
model.add(Dense(3, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y_onehot, epochs=300, batch_size=10)
```

```python
model.evaluate(X,  y_onehot, batch_size=10)
```

```python
model.predict_classes(X)
```
